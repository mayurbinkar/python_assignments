## Parms

```python
aUrl = 'https://api.openaq.org/v2/countries'

aParams = {'code': 'AF', 'name': 'Afghanistan', 'count': 23640, 'cities':1}

aResp = requests.get(aUrl, params=aParams)

print(aResp.json())
```

## Header

```python
import requests as rq
userAgent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'

checkHeader = rq.get('https://api.openaq.org/v2/countries', headers= {'User-Agent' : userAgent})

print(checkHeader)
```
